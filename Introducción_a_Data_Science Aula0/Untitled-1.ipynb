```python
# para ver el promedio de todas las peliculas
promedio_por_pelicula = notas.groupby('peliculaId').nota.mean()
promedio_por_pelicula
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    sns.histplot(data=promedio_por_pelicula, bins=20, kde=True)
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    plt.hist(promedio_por_pelicula, color='c')
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.legend(promedio_por_pelicula)
    plt.show()

promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    n, bins, patches = plt.hist(promedio_por_pelicula, color='c', alpha=0.7)  # alpha controla la opacidad de las barras
    for i in range(len(bins)-1):
        plt.text(bins[i] + (bins[i+1]-bins[i])/2, n[i], f'{int(n[i])}', ha='center', va='bottom', fontsize=12, color='b', fontweight='bold')
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    sns.boxplot(promedio_por_pelicula)
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.legend(promedio_por_pelicula, loc='best')
    plt.show()

promedio_pelicula()
```

```python
def idiomas_bar():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,5), facecolor='indigo')
    sns.barplot(data=contador_de_idiomas, x='lenguaje_original', y='total')
    plt.title('Peliculas',color='lightblue', size=20, fontweight='bold',pad=20)
    plt.xlabel('lenguaje_original',color='lightblue', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Idiomas',color='lightblue', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='lightblue', size=15, fontweight='bold')
    plt.yticks(color='lightblue', size=15, fontweight='bold')
    plt.show()
idiomas_bar()
```

```python
def idiomas_catplot():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,5), facecolor='indigo')
    sns.countplot(x=tmdb['lenguaje_original'], hue=tmdb['lenguaje_original'], palette='viridis', legend=False)
    plt.title('Peliculas',color='lightblue', size=20, fontweight='bold',pad=20)
    plt.xlabel('lenguaje_original',color='lightblue', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Cantidad de Pel√≠culas',color='lightblue', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='lightblue', size=15, fontweight='bold', rotation=45)  # Rotamos las etiquetas para mayor legibilidad
    plt.yticks(color='lightblue', size=15, fontweight='bold')
    plt.show()

idiomas_catplot()
```
