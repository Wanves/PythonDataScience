### Analizando algunas notas especificas por pelicula

```python
peliculas = pd.read_csv('movies.csv')
peliculas.head()
```

```python
# metodo para cambiar los t√≠tulos de las columnas
dict_pelicula = {
    'movieId':'peliculaId',
    'title':'t√≠tulo',
    'genres':'generos'
}
peliculas.rename(columns=dict_pelicula, inplace=True)
peliculas.head()
```

```python
notas.query('peliculaId == 1')
```

```python
notas.query('peliculaId == 1').nota.describe()
```

```python
# para ver el promedio de una pelicula
round(notas.query('peliculaId == 1').nota.mean(), 6)
```

```python
# para ver el promedio de todas las peliculas
promedio_por_pelicula = notas.groupby('peliculaId').nota.mean()
promedio_por_pelicula
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    sns.histplot(data=promedio_por_pelicula, bins=20, kde=True)
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    plt.hist(promedio_por_pelicula, color='c')
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.legend(promedio_por_pelicula)
    plt.show()

promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    n, bins, patches = plt.hist(promedio_por_pelicula, color='c', alpha=0.7)  # alpha controla la opacidad de las barras
    for i in range(len(bins)-1):
        plt.text(bins[i] + (bins[i+1]-bins[i])/2, n[i], f'{int(n[i])}', ha='center', va='bottom', fontsize=12, color='b', fontweight='bold')
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula()
```

```python
def promedio_pelicula():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    sns.boxplot(promedio_por_pelicula)
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.legend(promedio_por_pelicula, loc='best')
    plt.show()

promedio_pelicula()
```

```python
promedio_por_pelicula.describe()
```

```python
def promedio_pelicula_displot():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,8), facecolor='indigo')
    sns.distplot(promedio_por_pelicula, bins=10)
    plt.title('Datos de Pel√≠culas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Notas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula_displot()
```

```python
def promedio_pelicula_hist():
    sns.set(style="darkgrid")
    plt.figure(figsize=(20,5), facecolor='indigo')
    plt.hist(promedio_por_pelicula, color='c')
    n, bins, patches = plt.hist(promedio_por_pelicula, color='c', alpha=0.7)
    for i in range(len(bins)-1):
        plt.text(bins[i] + (bins[i+1]-bins[i])/2, n[i], f'{int(n[i])}', ha='center', va='bottom', fontsize=12, color='b', fontweight='bold')
    plt.title('Histograma Promedio peliculas',color='w', size=20, fontweight='bold',pad=20)
    plt.xlabel('Id Pel√≠culas',color='w', size=20, fontweight='bold',labelpad=20)
    plt.ylabel('Promedio',color='w', size=20, fontweight='bold',labelpad=20)
    plt.xticks(color='w', size=15, fontweight='bold')
    plt.yticks(color='w', size=15, fontweight='bold')
    plt.show()
promedio_pelicula_hist()
```
