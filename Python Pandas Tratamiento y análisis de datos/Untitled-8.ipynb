```python
seleccion_multiple = (datos['Tipo'] == 'Casa') | (datos['Tipo'] == 'Casa en condominio') | (datos['Tipo'] == 'Casa de villa')
seleccion_multiple
```

```python
df_seleccion_multiple = datos[seleccion_multiple]
df_seleccion_multiple.sort_index()
```

```python
total_casas = df_seleccion_multiple.shape[0]
```

```python
print(f'Los inmuebles casa casa , casa en condominio, casa de villa son en total: {total_casas}')
```

```python
seleccion = (datos['Area'] >= 60) & (datos['Area'] <= 100)
```

```python
df_area = datos[seleccion]
df_area.shape[0]
```

```python
print(f'Los inmuebles que cumplen la condición de medidas de Area son: {df_area.shape[0]}')
```
