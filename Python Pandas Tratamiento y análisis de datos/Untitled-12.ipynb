Creando nuevas variables

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial_sin_datos_nulos.csv', sep=';')
```

```python
datos.head()
```

```python
datos['Valor Bruto'] = (datos['Valor']) + (datos['Mantenimiento']) + (datos['Impuesto'])
datos['Valor Bruto'].sum()
```

```python
datos.head()
```

```python
datos['Valor m2'] = (datos['Valor']) / (datos['Area'])
datos.head()
```

```python
datos['Valor m2'] = round(datos['Valor m2'], 2)
datos.head()
```

```python
datos['Valor Bruto m2'] = round((datos['Valor Bruto']) / (datos['Area']), 2)
datos.head()
```

```python
casa = ['Casa', ' Casa en condominio', 'Casa de villa']
```

```python
datos['Tipo Agrupado'] = datos['Tipo'].apply(lambda x : 'Casa' if x in casa else 'Departamento')
datos
```

```python
datos.head()
```

```python
datos_aux = pd.DataFrame(datos[['Tipo Agrupado','Valor m2','Valor Bruto','Valor Bruto m2']])
```

```python
datos_aux.head()
```
