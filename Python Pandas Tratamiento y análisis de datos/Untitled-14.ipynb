Creando Agrupamientos

```python
import pandas as pd
```

```python
datos = pd.read_csv('alquiler_residencial_sin_datos_nulos.csv', sep=';')
```

```python
datos.head()
```

```python
datos['Valor'].mean()
```

```python
# barrios = ['San Borja','Lurigancho','Pachacámac','Ate','Comas']
# seleccion = datos['Distrito'].isin(barrios)
```

```python
# datos = datos[seleccion]
# datos.head()
```

```python
datos['Distrito'].drop_duplicates()
```

```python
grupo_barrio = datos.groupby('Distrito')
```

```python
grupo_barrio.groups
```

```python
for barrio, data in grupo_barrio:
    print(f'{barrio} -> {data.Valor.mean()} ')
```

```python
round(grupo_barrio[['Valor','Mantenimiento']].mean(), 2)
```

```python
round(grupo_barrio['Valor'].describe(), 2)
```

```python
grupo_barrio['Valor'].aggregate(['min', 'max']).rename(columns={
    'min':'Minimo',
    'max':'Máximo'
})
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].mean().plot.bar(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor medio del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].max().plot.bar(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor máximo del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].min().plot.bar(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor mínimo del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].std().plot.bar(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor desviación estandar del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].min().plot.hist(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor mínimo del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].min().plot.density(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor mínimo del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].min().plot.area(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor mínimo del Alquiler por distrito')
plt.show()
```

```python
plt.figure(figsize=(20,5))
fig = grupo_barrio['Valor'].min().plot.line(color='indigo')
fig.set_ylabel('Valor del Alquiler')
fig.set_title('Valor mínimo del Alquiler por distrito')
plt.show()
```

Creando rango de valor

```python
datos = pd.read_csv('alquiler_residencial_sin_datos_nulos.csv', sep=';')
```

```python
datos.head()
```

```python
clases = [0, 2, 4, 6, 100]
```

```python
cuartos = pd.cut(datos['Cuartos'], clases)
cuartos
```

```python
pd.Series(cuartos).value_counts()
```

```python
label = ['1 y 2 cuartos','3 y 4 cuartos','5 y 6 cuartos','7 o más cuartos']
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label )
cuartos
```

```python
cuartos = pd.cut(datos.Cuartos, clases, labels = label, include_lowest=True )
```

```python
cuartos
```

```python
pd.Series(cuartos).value_counts()
```

Cómo crear agrupaciones con groupby()
Estadísticas descriptivas con describe() y aggregate()

Cómo hacer gráficos con el paquete Matplotlib
Cómo crear rangos de valor con cut()
