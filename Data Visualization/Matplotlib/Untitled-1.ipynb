```python
temperatura_por_dia_semana = df.groupby('dia_de_semana')['temperatura'].mean()
nombre_dias = ['Domingo','Lunes', 'Martes', 'Miércoles', 'Jueves', 'Viernes', 'Sábado']
temperatura_por_dia_semana = temperatura_por_dia_semana[nombre_dias]
```

```python
fig = plt.figure(figsize=(15,5), facecolor='indigo')
eje = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_semana))
eje.tick_params(axis='x', colors='w')
eje.tick_params(axis='y', colors='w')
eje.bar(indice, temperatura_por_dia_semana)
eje.set_title('Temperatura diaria en una semana', fontsize=15,color='w',fontweight='bold', pad=10)
eje.set_xlabel('Día de la semana', fontsize=15, color='w',fontweight='bold')
eje.set_ylabel('Temperatura promedio', fontsize=15, color='w', fontweight='bold')
plt.show()
```

```python
fig = plt.figure(figsize=(15,5), facecolor='indigo')
eje = fig.add_axes([0,0,1,1])
indice = range(len(temperatura_por_dia_semana))
eje.tick_params(axis='x', colors='w')
eje.tick_params(axis='y', colors='w')
eje.bar(indice, temperatura_por_dia_semana)
eje.set_title('Temperatura diaria en una semana', fontsize=15,color='w',fontweight='bold', pad=10)
eje.set_xlabel('Día de la semana', fontsize=15, color='w',fontweight='bold')
eje.set_ylabel('Temperatura promedio', fontsize=15, color='w', fontweight='bold')
eje.set_xticks(indice)
eje.set_xticklabels(nombre_dias)



plt.show()
```
