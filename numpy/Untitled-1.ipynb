```python
años = np.loadtxt(fname='carros-años.txt')
km = np.loadtxt(fname='carros-km.txt')
valor = np.loadtxt(fname='carros-valor.txt')
```

```python
ds = np.column_stack((años, km, valor))
ds
```

```python
ds.shape
```

```python
media_kilometros = np.mean(ds[::,1])
media_kilometros
```

```python
media_valor = np.mean(ds[::,2])
round(media_valor, 2)
```

```python
np.mean(ds, axis = 0)
```

```python
media_kilometros = np.std(ds[::,1])
media_kilometros
```

```python
media_valor = np.std(ds[::,2])
round(media_valor, 2)
```
