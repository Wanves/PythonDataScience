```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
tmdb = pd.read_csv('tmdb_5000_movies.csv')
tmdb.head(2)
```

```python
tmdb.columns.unique()
```

```python
tmdb.describe()
```

```python
def grafico():
    sns.set(rc={"axes.facecolor": "lightblue", "figure.facecolor": "indigo"})
    ax = sns.displot(tmdb['vote_average'],kde=True,color='darkblue',  edgecolor='k');
    ax.figure.set_size_inches(20,7)
    plt.title('Grfico de distribucin del Promedio de notas Tmdb', fontsize=25, color='lightblue', fontweight='bold', pad=20)
    # ax.set_titles('Tmdb', fontsize=25, color='lightblue', fontweight='bold', pad=20)
    ax.set_xlabels('Nota Promedio', fontsize=25, color='lightblue', fontweight='bold', labelpad=20)
    ax.set_ylabels('Frecuencia',fontsize=25, color='lightblue', fontweight='bold', labelpad=20)
    plt.xticks(color='lightblue', size=15, fontweight='bold')
    plt.yticks(color='lightblue', size=15, fontweight='bold')
    plt.legend(tmdb['vote_average'])
    plt.show()
grafico()
```

```python
def grafico():
    sns.set(rc={"axes.facecolor": "lightblue", "figure.facecolor": "indigo"})
    ax = sns.boxplot(data=tmdb['vote_average'],color='indianred', orient='h', width=0.6);
    ax.figure.set_size_inches(20,7)
    plt.title('Grfico de boxplot del Promedio de Tmdb', fontsize=25, color='lightblue', fontweight='bold', pad=20)
    # ax.set_titles('Tmdb', fontsize=25, color='lightblue', fontweight='bold', pad=20)
    ax.set_xlabel('Nota Promedio', fontsize=25, color='lightblue', fontweight='bold', labelpad=20)
    ax.set_ylabel('Frecuencia',fontsize=25, color='lightblue', fontweight='bold', labelpad=20)
    plt.xticks(color='lightblue', size=15, fontweight='bold')
    plt.yticks(color='lightblue', size=15, fontweight='bold')
    plt.legend(tmdb['vote_average'])
    plt.show()
grafico()
```
