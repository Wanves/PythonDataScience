### 2. **Identificación de patrones**
   - **Agrupación por rendimiento:** Clasificaré a los preventistas según su desempeño en función de las variaciones en el volumen de ventas. Identificaré si alguno muestra una tendencia de crecimiento constante o si está teniendo una reducción significativa en su desempeño.

```python
data.sample()
```

```python
data_rendimiento = data.loc[:,['Nombre (preventista)','Volumen Act','% Var Volumen Mes']]
data_rendimiento
```

```python
data_rendimiento.dropna(inplace=True)
data_rendimiento
```

```python
# Clasificación basada en la variación de volumen de ventas
def clasificar_var(row):
    if row["% Var Volumen Mes"] > 2:
        return "Creciente"
    elif row["% Var Volumen Mes"] < -2:
        return "Decreciente"
    else:
        return "Constante"

data_rendimiento['Clasificación'] = data_rendimiento.apply(clasificar_var, axis=1)
data_rendimiento[['Nombre (preventista)', 'Volumen Act', '% Var Volumen Mes', 'Clasificación']].sort_values(by='Volumen Act', ascending=False)
```

```python
sns.set_style('darkgrid')
sns.set_theme(rc={'axes.facecolor':'lightblue', 'figure.facecolor':'indigo'})
plt.figure(figsize=(20,8))
plt.suptitle('Rendimiento ', color='lightblue', size=20, weight='bold', y=1.03)
plt.subplot(1,2,1)
ax = sns.barplot(x='Nombre (preventista)', y='Volumen Act', data=data_rendimiento, edgecolor='k', hue='Volumen Act', palette='Dark2')
for container in ax.containers:
    ax.bar_label(container,
                label_type='edge',
                padding=2,
                color='k',
                fontsize=12,
                fontweight='bold',
                fmt='{:,.0f}kg')
plt.title('Volumen Act ', color='lightblue', size=15, weight='bold', pad=20)
plt.xlabel('Nombre (preventista)', color='lightblue', size=15, weight='bold', labelpad=20)
plt.ylabel('Volumen Act', color='lightblue', size=15, weight='bold', labelpad=20)
plt.xticks(color='lightblue', size=10, weight='bold', rotation=45, ha='right')
plt.yticks(color='lightblue', size=10, weight='bold')
plt.legend().remove()
plt.subplot(1,2,2)
ax = sns.barplot(x='Nombre (preventista)', y='% Var Volumen Mes', data=data_rendimiento, edgecolor='k', hue='% Var Volumen Mes', palette='Dark2')
for container in ax.containers:
    ax.bar_label(container,
                label_type='edge',
                padding=2,
                color='k',
                fontsize=12,
                fontweight='bold',
                fmt='{:,.0f}%')
plt.title('% Var Volumen Mes', color='lightblue', size=15, weight='bold', pad=20)
plt.xlabel('Nombre (preventista)', color='lightblue', size=15, weight='bold', labelpad=20)
plt.ylabel('% Var Volumen Mes', color='lightblue', size=15, weight='bold', labelpad=20)
plt.xticks(color='lightblue', size=10, weight='bold', rotation=45, ha='right')
plt.yticks(color='lightblue', size=10, weight='bold')
plt.legend().remove()
plt.subplots_adjust(wspace=0.25)
plt.show()
```
