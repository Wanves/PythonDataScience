```python
plt.figure(figsize=(20, 5), facecolor='indigo')
sns.set(style="darkgrid")
sns.lineplot( data=x.head(20), x='score_gastos', y='ingreso');
plt.xlabel('score_gastos',color='w', size=20, fontweight='bold')
plt.ylabel('ingreso', color='w', size=20, fontweight='bold')
plt.xticks(fontweight='bold', fontsize=12, color='white')  # Haz que los números del eje x sean más gruesos
plt.yticks(fontweight='bold', fontsize=12, color='white')  # Haz que los números del eje y sean más gruesos
plt.legend(x['score_gastos'],bbox_to_anchor=(1.01, 1.02), loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(20, 5), facecolor='indigo');
sns.set(style="darkgrid");
sns.displot(x['ingreso'], height=5, aspect=4,color='indigo');
plt.xlabel('ingreso',color='k', size=20, fontweight='bold');
plt.ylabel('count', color='k', size=20, fontweight='bold');
plt.xticks(fontweight='bold', fontsize=12, color='k');  # Haz que los números del eje x sean más gruesos
plt.yticks(fontweight='bold', fontsize=12, color='k');  # Haz que los números del eje y sean más gruesos
plt.show();
```

```python
sns.set(style="darkgrid");
sns.displot(x['score_gastos'], height=5, aspect=4,color='indigo');
plt.xlabel('score_gastos',color='k', size=20, fontweight='bold')
plt.ylabel('count', color='k', size=20, fontweight='bold')
plt.xticks(fontweight='bold', fontsize=12, color='k')  # Haz que los números del eje x sean más gruesos
plt.yticks(fontweight='bold', fontsize=12, color='k')  # Haz que los números del eje y sean más gruesos
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
standart_scaler = StandardScaler()
```

```python
x_ss = standart_scaler.fit_transform(x)
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, random_state=0)
    kmeans.fit(x_ss)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(20, 5), facecolor='indigo')
sns.set(style="darkgrid")
plt.plot(range(1,11),wcss)
plt.xlabel(f'Número de clusters', color='w', size=15, fontweight='bold')
plt.ylabel(f'Inertia', color='w', size=15, fontweight='bold')
plt.xticks(fontweight='bold', fontsize=12, color='white')  # Haz que los números del eje x sean más gruesos
plt.yticks(fontweight='bold', fontsize=12, color='white')  # Haz que los números del eje y sean más gruesos
plt.show()
```

```python
kmeans = KMeans(n_clusters=5)
```

```python
df_cluster['Cluster'] = kmeans.fit_predict(x_ss)
```

```python
df_cluster.head()
```

```python
plt.figure(figsize=(20, 5), facecolor='indigo')
sns.set(style="darkgrid")
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 0,'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 0,'score_gastos'].values, c='r', label='Cluster 1')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 1,'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 1,'score_gastos'].values, c='b', label='Cluster 2')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 2,'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 2,'score_gastos'].values, c='orange', label='Cluster 3')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 3,'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 3,'score_gastos'].values, c='indigo', label='Cluster 4')
plt.scatter(x=df_cluster.loc[df_cluster.Cluster == 4,'ingreso'].values, y=df_cluster.loc[df_cluster.Cluster == 4,'score_gastos'].values, c='y', label='Cluster 5')
plt.xlabel('Ingreso ($)',fontweight='bold', color='w', size=20)
plt.ylabel('Score de Gastos', fontweight='bold', color='w', size=20)
plt.xticks(fontweight='bold', fontsize=12, color='white', size= 13)  # Haz que los números del eje x sean más gruesos
plt.yticks(fontweight='bold', fontsize=12, color='white', size= 13)  # Haz que los números del eje y sean más gruesos
plt.legend()
plt.show()
```
