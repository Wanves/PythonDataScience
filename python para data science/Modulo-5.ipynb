```python
import pandas as pd
import numpy as np
import matplotlib
from random import randrange, seed
import matplotlib.pyplot as plt
```

```python
notas = []
seed(8)
for i in range(6):
    notas.append(randrange(0,11))
```

```python
notas
```

```python
len(notas)
```

```python
x = list(range(1,7))
y = notas
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(10,5))
plt.plot(x, y, marker= 'o')
plt.title('Gráfico de Notas de matemáticas')
plt.xlabel('Periodo')
plt.ylabel('Notas')
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.barh(x, y)
plt.title('Gráfico de Notas de matemáticas')
plt.xlabel('Periodo')
plt.ylabel('Notas')
plt.show()
```
